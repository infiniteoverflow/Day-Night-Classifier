## Feature Extraction

```python
def average_brightness(image):
    hsv = cv2.cvtColor(image[0],cv2.COLOR_RGB2HSV)
    
    sum_brightness = np.sum(hsv[:,:,2])
    
    area = image[0].shape[0] * image[0].shape[1]
    
    return sum_brightness/area
```

```python
image_num = 1
test_im = image_list[image_num]

avg = average_brightness(test_im)
print('Avg brightness: ' + str(avg))
plt.imshow(test_im[0])
```

```python
def estimate_label(image):
    avg = average_brightness(image)
    
    threshold = 135
    
    predicted_label = 0
    
    if avg > threshold:
        predicted_label = 1
    
    return predicted_label
```

```python
predicted = []
misclassifed = []
correct_images = []
correct = 0
wrong = 0
for image in image_list:
    img = image[0]
    label = image[1]
    predicted = estimate_label(image)
    
    if label == predicted:
        correct += 1
        correct_images.append((img,label,predicted))
    else:
        wrong += 1
        misclassifed.append((img,label,predicted))

print("Accuracy : {:.2f} ".format((correct/(correct+wrong))))
```

```python
index = 28

print("Correct: ",misclassifed[index][1])
print("Predicted: ",misclassifed[index][2])
plt.imshow(misclassifed[index][0])
print("Avg Brightness: ",average_brightness(misclassifed[index]))
```
